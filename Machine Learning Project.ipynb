# Model Building

```python
df.head()
```

```python
x=df.iloc[:,:-1] #all rows and all columns except last
y=df.iloc[:,-1] #all rows and only last column
```

```python
# importing this library for spliting our data into training and testing
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1,stratify=y) #using stratify as to ensure that both the train and test sets have the proportion of examples in each class that is present in the provided “y” array
```

```python
# library for model selection
from sklearn.linear_model import LogisticRegression
```

```python
# with the help of this library we can check the accuracy of the model
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
```

```python
# creating a model for Logistic Regression and checking its accuracy
model=LogisticRegression()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)

train= model.score(xtrain,ytrain)
test = model.score(xtest,ytest)

print("train score =>",train)
print("test score =>",test)
```

```python
# libraries for model selection
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
# here we are creating objects for diffrent classification model
svm=SVC()
knn=KNeighborsClassifier()
```

```python
def func(model):
    model.fit(xtrain,ytrain)
    ypred=model.predict(xtest)
    
    train=model.score(xtrain,ytrain)
    test=model.score(xtest,ytest)
    print("Training Score =>",train)
    print("Testing Score =>",test)

    print(classification_report(ytest,ypred))
    return model
```

## SVM

```python
# Accuracy for Support vector machine
func(svm) #kernal="linear"
```

```python
k=["rbf","poly","sigmoid"]
for i in k:
    svm=SVC(kernel=i)
    svm.fit(xtrain,ytrain)
    ypred=svm.predict(xtest)
    print(f"{i} => {accuracy_score(ytest,ypred)}")
```

## KNN

```python
# Accuracy for K Nearest Neighbors 
func(knn)
```

```python
for i in range(1,30):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    ypred=knn.predict(xtest)
    print(f"{i}: {accuracy_score(ytest,ypred)}")
    
```

```python
knn=KNeighborsClassifier(n_neighbors=7)
func(knn)
```

## Decision Tree

```python
for i in range(1,50):
    dt=DecisionTreeClassifier(max_depth=i)
    dt.fit(xtrain,ytrain)
    ypred=dt.predict(xtest)
    print(f"{i}: {accuracy_score(ytest,ypred)}")
```

```python
dt1=DecisionTreeClassifier(max_depth=11)
func(dt)
#this model is overfitting as the training accuracy is 1 and testing accuracy is 0.79
```

```python
for i in range(1,50):
    dt2=DecisionTreeClassifier(min_samples_leaf=i)
    dt2.fit(xtrain,ytrain)
    ypred=dt2.predict(xtest)
    print(f"{i}: {accuracy_score(ytest,ypred)}")
```

```python
dt3=DecisionTreeClassifier(min_samples_leaf=5)
dt3.fit(xtrain,ytrain)
ypred=dt3.predict(xtest)
print(accuracy_score(ytest,ypred))

train=dt3.score(xtrain,ytrain)
test=dt3.score(xtest,ytest)

print("training score=>",train)
print("testing score=>",test)
```

## Decision Tree Regularization

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bg=BaggingClassifier(DecisionTreeClassifier(criterion="entropy"))
bg.fit(xtrain,ytrain)
ypred=bg.predict(xtest)

train=bg.score(xtrain,ytrain)
test=bg.score(xtest,ytest)

print("training score=>",train)
print("testing score=>",test)
```

```python
bg=BaggingClassifier(DecisionTreeClassifier())
bg.fit(xtrain,ytrain)
ypred=bg.predict(xtest)

train=bg.score(xtrain,ytrain)
test=bg.score(xtest,ytest)

print("training score=>",train)
print("testing score=>",test)
```
